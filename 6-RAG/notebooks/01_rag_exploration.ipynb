# RAG exploration notebook

This notebook is for experimenting with chunking, retrieval, and prompt design.

```python
from app.document_loader import load_text_files
from app.chunker import chunk_text
from app.retriever import retrieve_relevant_chunks

documents = load_text_files('data/raw/sample_docs')
chunks = []
for doc in documents:
    chunks.extend(chunk_text(doc))

results = retrieve_relevant_chunks('What is RAG?', chunks, top_k=3)
for chunk, score in results:
    print(score, chunk)
```
